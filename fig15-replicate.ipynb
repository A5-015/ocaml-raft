```python
graph_a = ['12-24','25-50','50-100','100-200','150-300']
graph_b = ['150-151','150-155','150-175','150-200','150-300']
```

```python
#read data in
data_dict = {}
for graph in [graph_a,graph_b]:
    for line in graph:
        file = open('data/'+line+'results.log')
        file_reader = csv.reader(file,delimiter='\n')
        data_dict[line] = []
        for point in file_reader:
            data_dict[line] += point
```

```python
# convert to ms floats
flt_dict = {}
for line in graph_a + graph_b:
    flt_dict[line] = []
    for point in data_dict[line]:
        flt_dict[line].append(float(point)/100.0)
    flt_dict[line].sort()
```

```python
y_axis = []
size = len(flt_dict[graph_a[0]])
for y in range (1,size+1):
    y_axis.append(y*100.0/size)
```

```python
for line in graph_a:
    if (len(flt_dict[line]) != len(y_axis)):
        print (line)
    plot(flt_dict[line],y_axis)
```
